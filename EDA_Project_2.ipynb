```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
```

```python
file_name = list(uploaded.keys())[0]
df = pd.read_excel(file_name)
```

```python
print("Dataset Shape:")
print(df.shape)
```

```python
print("\nColumns:")
print(df.columns)
```

```python
print("\nStatistical Summary:")
print(df.describe())
```

```python
print("\nAverage Total Price:")
print(df['TotalPrice'].mean())
```

```python
print("\nMedian Total Price:")
print(df['TotalPrice'].median())
```

```python
print("\nHighest Order Value:")
print(df['TotalPrice'].max())
```

```python
print("\nTop Selling Products:")
print(df['Product'].value_counts().head())
```

```python
print("\nMost Used Payment Method:")
print(df['PaymentMethod'].value_counts())
```

```python
high_orders = df[df['TotalPrice'] > df['TotalPrice'].mean() * 2]

print("Possible Outliers:")
print(high_orders[['OrderID', 'TotalPrice']].head())
```
